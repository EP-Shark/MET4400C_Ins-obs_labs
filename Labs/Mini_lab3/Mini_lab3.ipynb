```python
#Importing packages
import numpy as np
```

```python
# We will be calculating the height based off the pressure, and temperature measurements, using the ideal gas law.
# Ideal gas law for Delta Z = RdTavg/g * ln(P0/P)

#Data used:

#General values 
Rd = 287 # specific gas constant for dry air
g = 9.81 # Gravity in m/s
Po = 1013.25 #Average surface pressure of Earth

# Data For EOAS Building 
C_to_K = 273.15 # Celsius to kelvin conversion factor
Tsphere = 28.6 + C_to_K # degrees in kelvin
Troof = 26.3 + C_to_K # degrees K
Tav_r_s = np.mean([Tsphere, Troof])# Taking the average of temperature between the two locations

Psphere = 1021.2   # hPa (pressure)
Proof = 1018.1 # hPa (pressure)

# Data for around TLH

# Data from geodetic survey marker
Tg = 24 + C_to_K # Degrees K, acts as the reference temperature to be averaged
Pg = 1017.9 # hPa (pressure), geodetic survey marker acts as the reference point and consequently Po in hypsometric EQN
Zg = 67.230 # orthometric Height (z) of geodetic survey marker
#Data from lower point 
Tl2 = 23.4 + C_to_K # degrees K
Pl2 = 1021.1 # hPa (pressure)

#Data from higher point
Tl3 = 24.2 + C_to_K #Degree K
Pl3 = 1016.8 # hPa (pressure)

Tg_Tl2_avgT = np.mean([Tg, Tl2]) # Average temperature (K), between geodetic survey and location 2
Tg_Tl3_avgT = np.mean([Tg, Tl3]) # Average temperature (K), between geodetic survey and location 3
```

```python
# Calculating height of EOAS building data h_eos = z_roof - z_sphere = deltaZ

deltaZ= ((Rd * Tav_r_s)/g) * np.log(Psphere/Proof)

# Calculating the height difference between points around TLH
zl2 = Zg + ((Rd * Tg_Tl2_avgT)/g) * np.log(Pg/Pl2) # height at second location (orthometric like geodetic survey marker)
zl3 = Zg + ((Rd * Tg_Tl3_avgT)/g) * np.log(Pg/Pl3) # height at 3rd location (orthometric like geodetic survey marker)
```

```python
# Error propagation for height of EOAS building
SigmaP_kestrel = 0.1 # Error of the kestrel instrument in hPa
sigmaZ_eoas = ((Rd * Tav_r_s)/g) * SigmaP_kestrel * np.sqrt((1/Psphere)**2 + (1/Proof)**2)
sigmaZ_eoas 
print(f'The height of the EOAS building at FSU is {deltaZ:.2f} m \u00b1 {2 * sigmaZ_eoas:.2f} m at 95% confidence')
```

```python
sigmaZ_l2 = ((Rd * Tg_Tl2_avgT)/g) * SigmaP_kestrel * np.sqrt((1/Pg)**2 + (1/Pl2)**2)
sigmaZ_l2
sigmaZ_l3 = ((Rd * Tg_Tl2_avgT)/g) * SigmaP_kestrel * np.sqrt((1/Pg)**2 + (1/Pl3)**2)
sigmaZ_l3

print(f'The height of location 2 is {zl2:.2f} m \u00b1 {2 * sigmaZ_l2:.2f} m at 95% confidence')
print(f'The height of location 3 is {zl3:.2f} m \u00b1 {2 * sigmaZ_l3:.2f} m at 95% confidence')
```
